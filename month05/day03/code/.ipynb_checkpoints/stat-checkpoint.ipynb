## 协方差

```python
a = ratings['John Carson']
b = ratings['Michelle Peterson']
c = ratings['Melissa Jones']
# 计算协方差
a_mean = a.mean()
b_mean = b.mean()
c_mean = c.mean()
a_dev = a-a_mean
b_dev = b-b_mean
c_dev = c-c_mean
# print(c_dev)
cov_ab = np.mean(a_dev*b_dev)
cov_cb = np.mean(c_dev*b_dev)
# cov_cb
```

## 相关系数

```python
cor = cov_ab/(np.std(a)*np.std(b))
print(cor)
# 相关矩阵(每一个都是相关系数)
print(np.corrcoef((a,b,c)))
```
