```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 绘制简单直线
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 6, 9, 12, 14])
plt.plot(x, y)
plt.show()
```

```python
# 绘制水平线、垂线
plt.axhline(y=6, ls=":", c="blue")  # 添加水平直线
plt.axvline(x=4, ls="-", c="red")  # 添加垂直直线
plt.plot(x, y)
```

## 绘制正弦函数

```python
# -π - π 拆分出１０００个数
x = np.linspace(-np.pi,np.pi,1000)
# 求x的正弦值
y = np.sin(x)
plt.plot(x,y)
```

```python
cosx = np.cos(x)/2
plt.plot(x,cosx,linewidth=2,linestyle='-.',color='orangered',alpha=0.9)
plt.show()
```
